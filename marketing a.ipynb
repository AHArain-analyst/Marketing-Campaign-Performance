2. #### Context

3. #### Columns
   **Company**: The company responsible for the campaign, representing a mix of fictional brands.
   
   **Campaign_Type**: The type of campaign employed, including email, social media, influencer, display, or search.

   **Target_Audience**: The specific audience segment targeted by the campaign, such as women aged 25-34, men aged 18-24, or all age groups.

   **Duration**: The duration of the campaign, expressed in days.

   **Channels_Used**: The channels utilized to promote the campaign, which may include email, social media platforms, YouTube, websites, or Google Ads

   **Conversion_Rate**: The percentage of leads or impressions that converted into desired actions, indicating campaign effectiveness.

    **Acquisition_Cost**: The cost incurred by the company to acquire customers, presented in monetary format.

    **ROI**: Return on Investment, representing the profitability and success of the campaign.

    **Location**: The geographical location where the campaign was conducted, encompassing major cities like New York, Los Angeles, Chicago, Houston, or Miami.

    **Language**: The language used in the campaign communication, including English, Spanish, French, German, or Mandarin.

    **Clicks**: The number of clicks generated by the campaign, indicating user engagement.

    **Impressions**: The total number of times the campaign was displayed or viewed by the target audience.

    **Engagement_Score**: A score ranging from 1 to 10 that measures the level of engagement generated by the campaign.

    **Customer_Segment**: The specific customer segment or audience category that the campaign was tailored for, such as tech enthusiasts, fashionistas, health and wellness enthusiasts, foodies, or outdoor adventurers.

    **Date**: The date on which the campaign occurred, providing a chronological perspective to analyze trends and patterns.

# b4 Converting DataTypes
The data needs to be converted into their respective data types before analysis process begins

```python
def extract_gender_age(row):
    if 'All Ages' in row:
        gender = 'All'
        age = 'All'
    else:
        gender = row.split(' ')[0].strip()
        age = row.split(' ')[1].replace(' ', '').strip()
    return gender, age

# Apply the function to create new columns for gender and age
df['Gender'], df['Age'] = zip(*df['Target_Audience'].apply(extract_gender_age))

df.head()
```

```python
# filling out the place with most commen gender
df['Gender'] = df['Gender'].replace('All', pd.NA)
most_common_gender = df['Gender'].mode()[0]
df['Gender'] = df['Gender'].fillna(most_common_gender)

# filling out the place with most commen age
df['Age'] = df['Age'].replace('All', pd.NA)
most_common_age = df['Age'].mode()[0]
df['Age'] = df['Age'].fillna(most_common_age)

```

- Converting Data Types

```python
#Converting Duration to numeric values
df['Duration'] = pd.to_numeric(df['Duration'].astype(str).str.replace(' days', ''))

#Converting Company to categorical
df['Company'] = df['Company'].astype('category')

#Converting Campaign_Type to categorical
df['Campaign_Type'] = df['Campaign_Type'].astype('category')

#Converting Channel_Used to categorical
df['Channel_Used'] = df['Channel_Used'].astype('category')

#Converting Location to categorical
df['Location'] = df['Location'].astype('category')

#Converting Language to categorical
df['Language'] = df['Language'].astype('category')

#Converting Customer_Segment to categorical
df['Customer_Segment'] = df['Customer_Segment'].astype('category')

#Convert Gender to String
df['Gender'] = df['Gender'].astype('str')


print(df.dtypes)
df.head()

```

# 2. Channel Selection and Optimization:

## Problem Statement
- **Problem:** Unsure which channels are providing the best results.
- **Solution:** Evaluate the conversion rates and engagement scores for each channel. Optimize the budget allocation by focusing on channels that yield higher returns.

### 2.1. Calculate Average Conversion Rate and Engagement Score per Channel:

- Calculate the average conversion rate and engagement score for each channel.
- This will give you an overall idea of the performance of each channel.

```python
# Calculate average conversion rate and engagement score per channel
channel_performance = df.groupby('Channel_Used').agg({'Conversion_Rate': 'mean', 'Engagement_Score': 'mean'}).reset_index()

# Print the results
print("Average Conversion Rate and Engagement Score per Channel:")
channel_performance[['Channel_Used', 'Conversion_Rate', 'Engagement_Score']]
```

---

### 2.2. Identify High-Performing Channels:

- Identify channels with higher-than-average conversion rates and engagement scores.
- These channels are likely to be more effective in terms of generating conversions and engaging the audience.

```python
# Calculate average conversion rate and engagement score per channel
channel_performance = df.groupby('Channel_Used').agg({'Conversion_Rate': 'mean', 'Engagement_Score': 'mean'}).reset_index()

# Set your threshold for considering a channel as high-performing
conversion_rate_threshold = 0.01  # Adjust as needed
engagement_score_threshold = 1  # Adjust as needed

# Identify high-performing channels
high_performing_channels = channel_performance[
    (channel_performance['Conversion_Rate'] > conversion_rate_threshold) & 
    (channel_performance['Engagement_Score'] > engagement_score_threshold)
]

# Print the high-performing channels
print("High-Performing Channels:")
high_performing_channels[['Channel_Used', 'Conversion_Rate', 'Engagement_Score']]
```

#### Explanation:
- This code segment filters channels with conversion rates and engagement scores higher than the average.
- It helps identify channels that are more effective in generating conversions and engaging the audience.
- According to this the top 3 Highest Performing Channels are **Email, Facebook, and Google Ads**
---

### 2.3. Optimize Budget Allocation:

- Allocate a higher budget to the channels that have shown better performance.
- This doesn't mean completely discarding other channels but adjusting the budget distribution based on performance.

```python
# Calculate a composite performance score (e.g., weighted sum of Conversion Rate and Engagement Score)
df['Performance_Score'] = 0.6 * df['Conversion_Rate'] + 0.4 * df['Engagement_Score']

# Sort channels based on the performance score in descending order
df = df.sort_values(by='Performance_Score', ascending=False)

# Calculate the total budget
total_budget = df['Acquisition_Cost'].sum()

# Calculate the budget allocation percentage based on performance score
df['Budget_Allocation'] = (df['Performance_Score'] / df['Performance_Score'].sum()) * total_budget

# Display the optimized budget allocation
print("Optimized Budget Allocation:")
df[['Channel_Used', 'Budget_Allocation']].head(10)
```

#### Explanation:
- This code snippet calculates a composite performance score by assigning weights to Conversion Rate and Engagement Score.
- Channels are then sorted based on this score in descending order.
- The budget allocation percentage for each channel is calculated proportional to its performance score, and the total budget is distributed accordingly.
- The final result shows the optimized budget allocation for each channel.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Ensure all columns are of numeric type
df = df.apply(pd.to_numeric, errors='coerce')
```

```python

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[['Conversion_Rate', 'ROI']])
```

```python
df.head()
```

```python

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_data)

# Convert 'Cluster' column to numeric type
df['Cluster'] = df['Cluster'].astype('int')
```

```python
# Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Conversion_Rate', y='ROI', hue='Cluster', palette='viridis')
plt.title('K-means Clustering of Target Audience Segments')
plt.xlabel('Conversion Rate')
plt.ylabel('ROI')
plt.legend(title='Cluster')
plt.show()
```

```python
# Analyze the characteristics of each cluster
cluster_means = df.groupby('Cluster').mean()
print("Characteristics of Each Cluster:")
print(cluster_means)
```
